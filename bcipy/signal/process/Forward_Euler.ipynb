# Generate synthetic data with the model from arxiv

$$\frac{dx_{i}(t)}{dt} = -\frac{x_{i}}{\tau} + \sum_{j=1}^{N} {w_{ij}\sigma(x_{j},\theta)}+u(t)$$


$$V_{k}^{E} = M_k\bar{x}(t)$$

$$\frac{V_{k}^{E}}{dt} = M_k \frac{d\bar{x}(t)}{dt}$$

$$\frac{dV^{E}_{k}}{dt} = - \frac{M_k}{\tau}\bar{x} + M_{k}\sum_{j=1}^{N} {w_{ij}\sigma(x_{j},\theta)}+M_{k}u(t)$$

## Forward Euler

$$ x_i(t+\Delta{t}) = x_i(t) + \Delta{t}[-\frac{x_{i}}{\tau} + \sum_{j=1}^{N} {w_{ij}\sigma(x_{j},\theta)}+u(t)]$$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# Define the sigmoid function
def sigma(x, theta):
    return 1 / (1 + np.exp(-theta * x))

# Define the model differential equation for PyTorch
class ConnectivityModel(nn.Module):
    def __init__(self, N):
        super(ConnectivityModel, self).__init__()
        self.W = nn.Parameter(torch.randn(N, N))
        self.tau = 10.0
        self.theta = 1.0

    def forward(self, x, u):
        phi_x = 1 / (1 + torch.exp(-self.theta * x))
        dxdt = -x / self.tau + torch.matmul(phi_x, self.W.T) + u
        return dxdt

# Generate synthetic data
def generate_synthetic_data(w_true, input_signal, x0, t):
    def model_x(x, t, tau, w, theta, u):
        dxdt = np.zeros_like(x)
        N = len(x)
        time_index = min(int(t), len(u[0]) - 1)  # Ensure index is within bounds
        for i in range(N):
            dxdt[i] = -x[i] / tau + np.sum(w[i, :] * sigma(x, theta)) + u[i][time_index]
        return dxdt

    synthetic_data = []
    for _ in range(100):
        x = odeint(model_x, x0, t, args=(tau, w_true, theta, input_signal))
        synthetic_data.append(x)
    return np.array(synthetic_data)

# Set parameters
N = 3  # Number of neurons
tau = 10.0  # Time constant
theta = 1.0  # Sigmoid function parameter
w_true = np.random.randn(N, N)  # True weight matrix for generating synthetic data

# Initial conditions
x0 = np.random.randn(N)
input_signal = np.array(total_stimulis[0])  # Ensure this is your squared signal input of size (3, 792)

# Time vector
t = np.arange(0, 792)

# Generate synthetic data
synthetic_membrane_potentials = generate_synthetic_data(w_true, input_signal, x0, t)
print("Finished generating synthetic data")

# Convert synthetic data to numpy array and transpose to (100, 3, 792)
synthetic_membrane_potentials = np.array(synthetic_membrane_potentials).transpose(0, 2, 1)

# Convert data to PyTorch tensors
input_signal_torch = torch.tensor(input_signal, dtype=torch.float32)
synthetic_membrane_potentials_torch = torch.tensor(synthetic_membrane_potentials, dtype=torch.float32)

# Define the PyTorch model
model = ConnectivityModel(N)

# Define training parameters
learning_rate = 0.05
epochs = 100

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Function to solve the differential equation using PyTorch
def solve_differential_equation(model, x0, u, t):
    x_pred = [x0]
    x = x0
    for t_step in range(1, len(t)):
        u_t = u[:, t_step]
        dxdt = model(x, u_t)
        x = x + dxdt * (t[t_step] - t[t_step - 1])  # Using Euler's method for simplicity
        x_pred.append(x)
    return torch.stack(x_pred, dim=1)

# Training loop
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    
    loss = 0
    for i in range(10):
        x0_sample = synthetic_membrane_potentials_torch[i, :, 0]
        u_sample = input_signal_torch
        t_sample = torch.tensor(t, dtype=torch.float32)
        
        x_pred = solve_differential_equation(model, x0_sample, u_sample, t_sample)
        loss += criterion(x_pred, synthetic_membrane_potentials_torch[i])
    
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}')

# Extract learned connectivity matrix
learned_W = model.W.detach().numpy()
print("Learned connectivity matrix W:\n", learned_W)
print("True connectivity matrix W:\n", w_true)

# Plotting the learned vs true connectivity matrices
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
cax = axs[0].matshow(w_true, cmap='viridis')
fig.colorbar(cax, ax=axs[0])
axs[0].set_title('True Connectivity Matrix')

cax = axs[1].matshow(learned_W, cmap='viridis')
fig.colorbar(cax, ax=axs[1])
axs[1].set_title('Learned Connectivity Matrix')

plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.signal import iirnotch, filtfilt, butter, sosfiltfilt
from typing import Tuple

# Define the filters
class Notch:
    """Remove a single frequency"""

    def __init__(self, sample_rate_hz, remove_freq_hz=60.0, quality_factor=30):
        remove_freq_hz = remove_freq_hz / (sample_rate_hz / 2)
        self.b, self.a = iirnotch(remove_freq_hz, quality_factor)

    def __call__(self, data: np.ndarray, fs: int) -> Tuple[np.ndarray, int]:
        return filtfilt(self.b, self.a, data), fs


class Bandpass:
    """Preserve a specified range of frequencies"""

    def __init__(self, lo, hi, sample_rate_hz, order=5):
        nyq = 0.5 * sample_rate_hz
        lo, hi = lo / nyq, hi / nyq
        self.sos = butter(order, [lo, hi], analog=False, btype="band", output="sos")

    def __call__(self, data: np.ndarray, fs: int) -> Tuple[np.ndarray, int]:
        return sosfiltfilt(self.sos, data), fs

class Downsample:
    """Downsampling by an integer factor"""

    def __init__(self, factor: int = 2, *args, **kwargs):
        self.factor = factor

    def __call__(self, data: np.ndarray, fs: int) -> Tuple[np.ndarray, int]:
        return data[::self.factor], fs // self.factor

class Composition:
    """Applies a sequence of transformations"""

    def __init__(self, *transforms):
        self.transforms = transforms

    def __call__(self, data: np.ndarray, fs: int) -> Tuple[np.ndarray, int]:
        for transform in self.transforms:
            data, fs = transform(data, fs)
        return data, fs

def get_default_transform(
    sample_rate_hz: int,
    notch_freq_hz: int,
    bandpass_low: int,
    bandpass_high: int,
    bandpass_order: int,
    downsample_factor: int,
    notch_quality_factor: int = 30,
) -> Composition:
    return Composition(
        Notch(sample_rate_hz, notch_freq_hz, notch_quality_factor),
        Bandpass(bandpass_low, bandpass_high, sample_rate_hz, bandpass_order),
        Downsample(downsample_factor),
    )

# Function to process the data
def process_eeg_data(triggers_file, eeg_file):
    # Read the triggers file
    with open(triggers_file, 'r') as file:
        lines = file.readlines()
    starting_offset = abs(float(lines[0].split()[-1]))

    # Read the EEG data
    eeg_data = pd.read_csv(eeg_file, skiprows=2)

    # Filter the rows based on the starting offset
    filtered_data = eeg_data[eeg_data['lsl_timestamp'] >= starting_offset]

    # Select only the sensors called Cz, O1, O2
    selected_data = filtered_data[['lsl_timestamp', 'Cz', 'O1', 'O2']].copy()

    # Compute the relative time between rows
    initial_lsl_timestamp = selected_data['lsl_timestamp'].iloc[0]
    selected_data['relative_time'] = selected_data['lsl_timestamp'] - initial_lsl_timestamp

    # Define transformation parameters
    sample_rate_hz = 300  # Assuming a sample rate of 250 Hz
    transform = get_default_transform(
        sample_rate_hz=sample_rate_hz,
        notch_freq_hz=60,
        bandpass_low=1,
        bandpass_high=40,
        bandpass_order=5,
        downsample_factor=2
    )

    # Apply filters and downsampling
    processed_signals = {}
    for sensor in ['Cz', 'O1', 'O2']:
        signal = selected_data[sensor].values
        signal, _ = transform(signal, sample_rate_hz)
        processed_signals[sensor] = signal

    # Determine the new length after downsampling
    downsampled_length = len(processed_signals['Cz'])

    # Create a new DataFrame with the downsampled signals and relative time
    processed_data = pd.DataFrame({
        'relative_time': selected_data['relative_time'].values[:downsampled_length],
        'Cz': processed_signals['Cz'],
        'O1': processed_signals['O1'],
        'O2': processed_signals['O2']
    })

    # Calculate the time in steps
    step_size = 1.0 / (sample_rate_hz / 2)  # Downsample factor of 2
    processed_data['time_steps'] = np.arange(0, downsampled_length) * step_size

    # Create the stimuli signal as a spike train
    stimuli_signal = np.zeros(len(filtered_data))
    lsl_timestamps = filtered_data['lsl_timestamp'].values

    for i, line in enumerate(lines[1:]):
        parts = line.split()
        if len(parts) == 3 and parts[1] not in ['prompt', 'fixation']:
            label, category, timestamp = parts
            timestamp = float(timestamp)
            if timestamp >= starting_offset:
                # Find the closest index in the EEG data
                idx = np.searchsorted(lsl_timestamps, timestamp)
                if idx < len(stimuli_signal):
                    if category=='target':
                        stimuli_signal[idx-40 : idx+40] = 1
                    elif category == 'nontarget':
                        stimuli_signal[idx-40 : idx+40] = 0.5

    # Create a DataFrame for the stimuli signal
    stimuli_df = pd.DataFrame({
        'lsl_timestamp': lsl_timestamps,
        'stimuli': stimuli_signal
    })
    stimuli_df['relative_time'] = stimuli_df['lsl_timestamp'] - initial_lsl_timestamp
    return processed_data, stimuli_df

# Example usage
triggers_file = '/Users/aliag/Desktop/Data/S002/S002_Matrix_Calibration_Thu_18_May_2023_12hr43min40sec_-0400/triggers_2.txt'
eeg_file =  '/Users/aliag/Desktop/Data/S002/S002_Matrix_Calibration_Thu_18_May_2023_12hr43min40sec_-0400/raw_data_2.csv'
processed_data, stimuli_df = process_eeg_data(triggers_file, eeg_file)

print(processed_data)
print(stimuli_df)



```

```python
def divide_stimuli_by_inquiries(triggers_file, stimuli_df):
    # Read the triggers file
    with open(triggers_file, 'r') as file:
        lines = file.readlines()

    # Identify the fixation timestamps and the first letter timestamps
    fixation_timestamps = []
    first_letter_shown = None

    for line in lines:
        parts = line.split()
        if len(parts) == 3:
            label, category, timestamp = parts
            if category == 'fixation':
                fixation_timestamps.append(float(timestamp))
            if category not in ['prompt', 'fixation'] and first_letter_shown is None:
                first_letter_shown = float(timestamp)

    # Adjust the start time of the first inquiry to the timestamp of the first letter shown
    if first_letter_shown is not None:
        fixation_timestamps[0] = first_letter_shown

    # Add the last timestamp of stimuli_df to cover the last inquiry
    fixation_timestamps.append(stimuli_df['lsl_timestamp'].iloc[-1] + 1)  # Adding 1 to ensure inclusion

    # Divide the stimuli_df by inquiries
    divided_stimuli = []
    for i in range(len(fixation_timestamps) - 1):
        start_time = fixation_timestamps[i]
        end_time = fixation_timestamps[i + 1]
        inquiry_df = stimuli_df[(stimuli_df['lsl_timestamp'] >= start_time) & (stimuli_df['lsl_timestamp'] < end_time)].copy()
        divided_stimuli.append(np.array(inquiry_df['stimuli']))

    return divided_stimuli


# Example usage
triggers_file = '/Users/aliag/Desktop/Data/S002/S002_Matrix_Calibration_Thu_18_May_2023_12hr43min40sec_-0400/triggers_2.txt'
divided_stimuli = divide_stimuli_by_inquiries(triggers_file, stimuli_df)

# Verify the result
for i, inquiry_df in enumerate(divided_stimuli):
    print(f"Inquiry {i+1}:")
    print(inquiry_df)
    print("Length:", len(inquiry_df))
    print()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import iirnotch, filtfilt, butter, sosfiltfilt
from typing import Tuple

# Define the RNN model
class RNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(RNNModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)  # Initialize hidden state
        out, _ = self.rnn(x, h0)  # Forward pass
        out = self.fc(out[:, -1, :])  # Take the last time step's output
        return out

# Filters
class Notch:
    def __init__(self, sample_rate_hz, remove_freq_hz=60.0, quality_factor=30):
        remove_freq_hz = remove_freq_hz / (sample_rate_hz / 2)
        self.b, self.a = iirnotch(remove_freq_hz, quality_factor)

    def __call__(self, data: np.ndarray, fs: int) -> Tuple[np.ndarray, int]:
        return filtfilt(self.b, self.a, data.copy()), fs

class Bandpass:
    def __init__(self, lo, hi, sample_rate_hz, order=5):
        nyq = 0.5 * sample_rate_hz
        lo, hi = lo / nyq, hi / nyq
        self.sos = butter(order, [lo, hi], analog=False, btype="band", output="sos")

    def __call__(self, data: np.ndarray, fs: int) -> Tuple[np.ndarray, int]:
        return sosfiltfilt(self.sos, data.copy()), fs

class Downsample:
    def __init__(self, factor: int = 2, *args, **kwargs):
        self.factor = factor

    def __call__(self, data: np.ndarray, fs: int) -> Tuple[np.ndarray, int]:
        return data[::self.factor].copy(), fs // self.factor

class Composition:
    def __init__(self, *transforms):
        self.transforms = transforms

    def __call__(self, data: np.ndarray, fs: int) -> Tuple[np.ndarray, int]:
        for transform in self.transforms:
            data, fs = transform(data, fs)
        return data, fs

def get_default_transform(sample_rate_hz: int, notch_freq_hz: int, bandpass_low: int, bandpass_high: int, bandpass_order: int, downsample_factor: int, notch_quality_factor: int = 30) -> Composition:
    return Composition(
        Notch(sample_rate_hz, notch_freq_hz, notch_quality_factor),
        Bandpass(bandpass_low, bandpass_high, sample_rate_hz, bandpass_order),
        Downsample(downsample_factor),
    )

# Custom loss function that applies the filters
class FilteredMSELoss(nn.Module):
    def __init__(self, sample_rate_hz, notch_freq_hz, bandpass_low, bandpass_high, bandpass_order, downsample_factor, notch_quality_factor=30):
        super(FilteredMSELoss, self).__init__()
        self.transform = get_default_transform(sample_rate_hz, notch_freq_hz, bandpass_low, bandpass_high, bandpass_order, downsample_factor, notch_quality_factor)
        self.mse_loss = nn.MSELoss()

    def forward(self, output, target):
        output_np = output.cpu().detach().numpy().copy()
        target_np = target.cpu().detach().numpy().copy()
        output_filtered, _ = self.transform(output_np, sample_rate_hz)
        target_filtered, _ = self.transform(target_np, sample_rate_hz)
        output_filtered = torch.tensor(output_filtered, dtype=torch.float32).to(output.device)
        target_filtered = torch.tensor(target_filtered, dtype=torch.float32).to(target.device)
        return self.mse_loss(output_filtered, target_filtered)

# Parameters
input_size = 1
hidden_size = 128
num_layers = 1
output_size = 1
num_epochs = 5
batch_size = 16
learning_rate = 0.01
sample_rate_hz = 250  # Assuming a sample rate of 250 Hz

# Replace these with your actual data tensors
x_data = np.random.randn(1, 315457)  # Synthetic membrane potentials
VE_data = np.random.randn(1, 315457)  # Synthetic raw time series EEG

# Convert data to PyTorch tensors and reshape
data = torch.tensor(x_data, dtype=torch.float32).unsqueeze(-1)  # (1, 315457, 1)
labels = torch.tensor(VE_data, dtype=torch.float32).unsqueeze(-1)  # (1, 315457, 1)

print("Data shape:", data.shape)
print("Labels shape:", labels.shape)

# Prepare the dataset and DataLoader
dataset = TensorDataset(data, labels)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Initialize the model, loss function, and optimizer
model = RNNModel(input_size, hidden_size, num_layers, output_size)
criterion = FilteredMSELoss(sample_rate_hz, 60, 1, 40, 5, 2)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    model.train()  # Set the model to training mode
    for batch_data, batch_labels in dataloader:
        # Forward pass
        outputs = model(batch_data)
        loss = criterion(outputs.view(-1, 315457), batch_labels.view(-1, 315457))
        
        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Evaluation
model.eval()
with torch.no_grad():
    total_loss = 0
    predictions = []
    actuals = []
    for batch_data, batch_labels in dataloader:
        outputs = model(batch_data)
        loss = criterion(outputs.view(-1, 315457), batch_labels.view(-1, 315457))
        total_loss += loss.item()
        predictions.append(outputs)
        actuals.append(batch_labels)
    
    average_loss = total_loss / len(dataloader)
    print(f'Average Evaluation Loss: {average_loss:.4f}')

# Concatenate all predictions and actuals for plotting
predictions = torch.cat(predictions).cpu().numpy()
actuals = torch.cat(actuals).cpu().numpy()

# Plot the predictions vs the corresponding labels
plt.figure(figsize=(10, 6))
plt.plot(predictions[0], label='Predicted', color='b')
plt.plot(actuals[0], label='Actual', color='r')
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Prediction vs Actual')
plt.legend()
plt.show()

```
